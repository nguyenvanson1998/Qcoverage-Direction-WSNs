```python
import random
import numpy as np
```

# Group 1
- Change the number of sensor from 20 to 180

```python
AREA = [500, 500]
noTarget = 130
minSensor =  20
upperSensor = 200
steps = 20
R = 100
phi = 90
```

```python
targets = np.random.uniform(low=[0,0], high=AREA, size=(130,2))
```

```python
no_q = np.random.randint(low =1, high=5, size=(130,1))
```

```python
targets = np.concatenate((targets, no_q),axis = 1)
sensors = np.random.uniform(low=[0,0], high=AREA, size=(180,2))
```

```python
for i in range(minSensor, upperSensor, steps):
    with open(f"../data/test_gr1_{i}sensor.inp", mode='w+') as fin:
        fin.writelines(f"{phi}\n")
        fin.writelines(f"{R}\n")
        fin.writelines(f"{AREA[0]},{AREA[1]}\n")
        fin.writelines(f"{noTarget}\n")
        fin.writelines(f"{i}\n")
        fin.writelines("\n".join(f"{t[0]},{t[1]},{t[2]}" for t in targets))
        fin.writelines("\n")
        #sensors = np.random.uniform(low=[0,0], high=AREA, size=(i,2))
        sensor1 = sensors[:i]
        fin.writelines("\n".join(f"{s[0]},{s[1]}" for s in sensor1))
        
```
